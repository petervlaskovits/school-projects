# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bank_marketing = pd.read_csv('data/bank-additional-full.csv', sep=';')
```

```python
ax = sns.histplot(
    data=bank_marketing,
    x='age',
    kde=False,
    stat='density',
    color='gray'
)

sns.kdeplot(
    data=bank_marketing,
    x='age',
    color='crimson',
    ax=ax
);
```
