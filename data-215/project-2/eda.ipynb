# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bank_marketing = pd.read_csv('data/bank-additional-full.csv', sep=';')
bank_marketing.head()
```

```python
bank_marketing.isna().mean()
```

```python
bank_marketing.isin(['unknown']).mean()
```

## Bank Client

### Data Dictionary

* Age

* Job

* Marital status

* Education

* Default

* Housing loan

* Loan

### Data visualization

```python
client_cols = [
    "age",
    "job",
    "marital",
    "education",
    "default",
    "housing",
    "loan",
]

client_data = bank_marketing[client_cols]
client_data.head()
```

```python
ax = sns.histplot(
    data=client_data,
    x='age',
    kde=False,
    stat='density',
    color='gray'
)

sns.kdeplot(
    data=client_data,
    x='age',
    color='crimson',
    linewidth=4,
    ax=ax
);
```

```python
sns.countplot(
    data=client_data,
    x='job',
    stat='percent'
)

plt.xticks(rotation=45);
```

```python
sns.countplot(
    data=client_data,
    x='marital',
    stat='percent'
);
```

```python
sns.countplot(
    data=client_data,
    x='education',
    stat='percent'
)

plt.xticks(rotation=45);
```

```python
sns.countplot(
    data=client_data,
    x='default',
    stat='percent'
);
```

```python
sns.countplot(
    data=client_data,
    x='housing',
    stat='percent'
);
```

```python
sns.countplot(
    data=client_data,
    x='loan',
    stat='percent'
);
```
