# Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

bank_marketing = pd.read_csv('data/raw/bank-additional-full.csv', sep=';')
bank_marketing.head()
```

```python
bank_marketing.isna().mean()
```

```python
bank_marketing.isin(['unknown']).mean()
```

## Bank Client

### Data Dictionary

* Age - age of the client

* Job - what type of job they have

* Marital status - their marital status

* Education - what type of education they have

* Default - Has the client defaulted on credit?

* Housing loan - Does the client have a home loan?

* Loan - Does the client have a personal loan? 

### Data visualization

```python
client_cols = [
    "age",
    "job",
    "marital",
    "education",
    "default",
    "housing",
    "loan",
]

client_data = bank_marketing[client_cols]
client_data.head()
```

```python
ax = sns.histplot(
    data=client_data,
    x='age',
    kde=False,
    stat='density',
    color='gray'
)

graph = sns.kdeplot(
    data=client_data,
    x='age',
    color='crimson',
    linewidth=4,
    ax=ax
);

graph.set_title('Distribution of age among clients')
graph.set_xlabel('Age');
```

```python
graph = sns.countplot(
    data=client_data,
    x='job',
    stat='percent'
)

plt.xticks(rotation=45)
graph.set_title("Jobs among clients")
graph.set_xlabel("Type of job")
graph.set_ylabel("Percent of clients");
```

```python
graph = sns.countplot(
    data=client_data,
    x='marital',
    stat='percent'
);

graph.set_title("Maritial status among clients")
graph.set_xlabel("Maritial status")
graph.set_ylabel("Percent of clients");
```

```python
graph = sns.countplot(
    data=client_data,
    x='education',
    stat='percent'
)

plt.xticks(rotation=45);
graph.set_title("Education level among clients")
graph.set_xlabel("Education level")
graph.set_ylabel("Percent of clients");
```

```python
graph = sns.countplot(
    data=client_data,
    x='default'
)

graph.set_title("Number of clients that have defaulted on a loan")
graph.set_xlabel("Defaulted on a loan?")
graph.set_ylabel("Number of clients");
```

```python
graph = sns.countplot(
    data=client_data,
    x='housing',
    stat='percent'
)

graph.set_title("Percentage of clients that have a home loan")
graph.set_xlabel("Has a home loan?")
graph.set_ylabel("Percent of clients");
```

```python
graph = sns.countplot(
    data=client_data,
    x='loan',
    stat='percent'
);

graph.set_title("Percentage of clients that have a personal loan")
graph.set_xlabel("Has a personal loan?")
graph.set_ylabel("Percent of clients");
```

## Last Contact

```python
last_contact_cols = [
    "contact",
    "month",
    "day_of_week",
    "duration"
]

last_contact = bank_marketing[last_contact_cols]
last_contact.head()
```

### Data Dictionary

* Contact - What was the client previously contacted by?
* Month - The month of when they were previously contacted
* Day of week - The day of when they were previously contacted
* Duration - How long did the call last?

### Data Visualization

```python
graph = sns.countplot(
    data=last_contact,
    x='contact',
    stat='percent'
);

graph.set_title("Contact methods when previously contacted")
graph.set_xlabel("Contact method")
graph.set_ylabel("Percent of clients previously contacted");
```

```python
graph = sns.countplot(
    data=last_contact,
    x='month',
    stat='percent'
);

graph.set_title("Month of contact when previously contacted")
graph.set_xlabel("Month")
graph.set_ylabel("Percent of clients previously contacted");
```

```python
graph = sns.countplot(
    data=last_contact,
    x='day_of_week',
    stat='percent'
);

graph.set_title("Day of contact when contacted")
graph.set_xlabel("Day")
graph.set_ylabel("Percent of clients previously contacted");
```

```python
graph = sns.histplot(
    data=last_contact,
    x='duration'
);

graph.set_title("Duration of call when previously contacted")
graph.set_xlabel("Call duration (seconds)")
graph.set_ylabel("Count");
```

```python
graph = sns.histplot(
    data=last_contact,
    x='duration',
    log_scale=True
);

graph.set_title("Duration of call when previously contacted (log scale)")
graph.set_xlabel("Call duration (seconds)")
graph.set_ylabel("Count");
```

## Other

```python
other_cols = [
    'campaign',
    'pdays',
    'previous',
    'poutcome'
]

other = bank_marketing[other_cols]
other.head()
```

### Data Dictionary

* Campaign - # of contacts during the campaign for the client (includes last contact)
* Pdays - Days that passed by after client got contacted by last campaign
* Previous - Number of contacts performed before this campaign and for this client
* Poutcome - Previous marketing campaign outcome

### Data Visualizations

```python
graph = sns.countplot(data=other, x='campaign');
plt.xticks(rotation=45);

graph.set_title("Number of contacts during campaign")
graph.set_xlabel("Number of contacts")
graph.set_ylabel("Count");
```

```python
graph = sns.histplot(data=other, x='pdays');
graph.set_title("Days since previously contacted (999 = not contacted at all)")
graph.set_xlabel("Days")
graph.set_ylabel("Count");
```

```python
graph = sns.countplot(data=other, x='previous');
graph.set_title("Number of contacts performed before the campaign, for the client")
graph.set_xlabel("Contacts")
graph.set_ylabel("Count");
```

```python
graph = sns.countplot(data=other, x='poutcome');
graph.set_title("Previous outcome")
graph.set_xlabel("Outcome")
graph.set_ylabel("Number of clients");
```

## Socioeconomic Factors

```python
socioeconomic_cols = [
    'emp.var.rate',
    'cons.price.idx',
    'cons.conf.idx',
    'euribor3m',
    'nr.employed'
]

socioeconomic = bank_marketing[socioeconomic_cols]
socioeconomic.head()
```

### Data Dictionary

* emp.var.rate - Employment variation rate in Portugal, quarterly
* cons.price.idx - Consumer price index in Portugal, monthly
* cons.conf.idx - Consumer confidence index in Portugal, monthly
* euribor3m - Interest rate in Portugal, 3 month rate, daily
* nr.employed - Number of employees employed, quarterly
